```python
def binary(num):
    return ''.join(bin(c).replace('0b', '').rjust(8, '0') for c in struct.pack('!f', num))
```

```python
def make_binary_list(target_list):
    a = 1
    while True:
        if 2 ** (a-1) > max(target_list):
            break
        else:
            a += 1
    target_binary = []
    for i in target_list:
        tmp = bin(i)[2:]
        tmp2 = ""
        for j in range(a-len(tmp)):
            tmp2+="0"
        tmp = tmp2 + tmp
        target_binary.append(tmp)
    return target_binary
```

```python
id_binary = make_binary_list(df_sample.id)
longtitude_binary = make_binary_list(df_sample.longtitude)
latitude_binary = make_binary_list(df_sample.latitude)
```

```python
id_longtitude_binary = []
for i in range(len(id_binary)):
    id_longtitude_binary.append([id_binary[i]+longtitude_binary[i]])
id_longtitude_binary = np.array(id_longtitude_binary)
```

```python
id_latitude_binary = []
for i in range(len(id_binary)):
    id_latitude_binary.append([id_binary[i]+latitude_binary[i]])
id_latitude_binary = np.array(id_latitude_binary)
```

```python
id_longtitude_binary_value = np.hstack([id_longtitude_binary,np.array([[int(i[0],2)] for i in id_longtitude_binary])])
id_latitude_binary_value = np.hstack([id_latitude_binary,np.array([[int(i[0],2)] for i in id_latitude_binary])])
```

```python
float_value_long = []
for i in id_longtitude_binary_value:
    float_value_long.append([np.float32(np.float32((int(i[1])/2**len(i[0]))) * 3.402823e+38),binary(np.float32(np.float32((int(i[1])/2**len(i[0]))) * 3.402823e+38))])
float_value_long = np.array(float_value_long)
```

```python
float_value_la = []
for i in id_latitude_binary_value:
    float_value_la.append([np.float32(np.float32((int(i[1])/2**len(i[0]))) * 3.402823e+38),binary(np.float32(np.float32((int(i[1])/2**len(i[0]))) * 3.402823e+38))])
float_value_la = np.array(float_value_la)
```

```python
id_longtitude_binary_value_float = np.hstack([id_longtitude_binary_value,float_value_long])
id_latitude_binary_value_float = np.hstack([id_latitude_binary_value,float_value_la])
```

```python
def float_to_num(target):
    for i in range(0,len(target),8):
        if i == 0:
            tmp = struct.pack("!B",int(target[i:i+8],2))
        else:
            tmp = tmp + struct.pack("!B",int(target[i:i+8],2))
    return np.float32(struct.unpack("!f",tmp)[0])
```
